```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline 
```

```python
from sklearn.datasets import load_iris 
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2, random_state = 5)
```

```python
## apply decision tree classifier 
from sklearn.tree  import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth = 2)# max depth is the feature through which we use post prunning
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
# visualization 
from sklearn import tree 
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled = True)
plt.show()
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report 
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
